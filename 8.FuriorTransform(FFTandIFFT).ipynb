```python
# FFT and IFFT

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1-D FFT
t = np.arange(128)
print(t)

sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(t.shape[-1])
```

```python
print(sp, freq)
```

```python
plt.subplot(2, 1, 1)
plt.title("Real Part")
plt.plot(freq, sp.real)

plt.subplots_adjust(hspace=0.5)

plt.subplot(2, 1, 2)
plt.title('Imaginary part')
plt.plot(freq, sp.imag)

plt.show()
```

```python
# Inverse 1-D DFT

t = np.arange(500)
s = np.zeros((500,), dtype = complex)

s[40:80] = np.exp(1j*np.random.uniform(0, 2*np.pi, (40,)))
n = np.fft.ifft(s)
# print(t, s)

plt.subplot(2, 1, 1)
plt.title("Real Part")
plt.plot(t, n.real)

plt.subplots_adjust(hspace=0.5)

plt.subplot(2, 1, 2)
plt.title('Imaginary part')
plt.plot(t, n.imag)

plt.show()
```

```python
plt.plot(t, n.real)
plt.title('1-D IFFT')
plt.plot(t, n.imag)

plt.show()
```
